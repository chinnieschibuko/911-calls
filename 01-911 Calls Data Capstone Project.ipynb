## Creating new features

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df,palette="magma")
```

```python
type(df["timeStamp"].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time : time.hour) 
df['Month'] = df['timeStamp'].apply(lambda time : time.month) 
df['Day of Week'] = df['timeStamp'].apply(lambda time : time.dayofweek) 
```

```python
df['Day of Week']
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df['Day of Week']
```

```python
sns.countplot(x = 'Day of Week', data = df, hue ='Reason' )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x = 'Month', data = df, hue ='Reason' )
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
count_month = df.groupby('Month').count()
count_month.head()
```

```python
count_month['lat'].plot()
```

```python
sns.lmplot(x='Month',y='lat',data=count_month.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda date : date.date())
df['Date']
```

```python
df.groupby('Date').count()['lat'].plot()
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
day_Hour = df.groupby(by=['Day of Week','Hour']).count()['lat'].unstack()
day_Hour.head()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(day_Hour,cmap='magma')
```

```python
plt.figure(figsize = (20,10))
sns.clustermap(day_Hour,cmap='magma')
```

```python
day_Month = df.groupby(by=['Day of Week','Month']).count()['lat'].unstack()
day_Month.head()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(day_Month,cmap='magma')
```
